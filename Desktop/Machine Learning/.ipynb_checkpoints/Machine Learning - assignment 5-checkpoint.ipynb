```python
import pandas as pd

df = pd.read_excel("M3 Raisin_Dataset.xlsx")
df.head()
```

**explanation**

- load dataset and show the first few rows

```python
# Check for any missing variables.
df.isnull().sum()
```

```python
# duplicated rows
duplicated = df.duplicated().sum()
print(f"Duplicate rows: {duplicated}")
```

**explanation**

- checked for null values and duplicated rows and there were none

```python
df.info()
```

**explanation**

- check for each feature's type

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Class'] = le.fit_transform(df['Class'])  
```

```python
df.head()
```

**explanation**

- model will not be able to read class, so used label encoder to transform Kecimen to 1 and Besni to 0

```python
# scaling
from sklearn.preprocessing import StandardScaler

X = df.drop('Class', axis=1)
y = df['Class']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**explanation**

- Assigned feature (X) and Target (Y)
- Target is the class, and the rest are features (X)
- looking at the features, it will need to be scaled
- scaling with standard scaler (the heatmap below will show the scaled value for each feature)

```python
# plot 1
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

**explanation**

- the first plot shows how features correlate with each other
- As shown in the plot, "Area" and "ConvexArea" are highly correlated, and we should drop one to avoid multicollinearity.

```python
df = df.drop('ConvexArea', axis=1)
```

**explanation**

- droipping "Convexarea"

```python
# plot 2
plt.figure()
sns.boxplot(x='Class', y='Area', data=df)
plt.title("Boxplot of Area by Class")
plt.show()
```

**explanation**

- the box plot shows class 0 has a higher median than class 1 (Raisins in class 0 is larger)
- class 1 has a smaller spread of raisin size
- both classes has outliers (some larger raisins)

```python
# plot 3
sns.histplot(data=df, x='Area', hue='Class', kde=True)
plt.title("Distribution of Area by Class")
plt.show()
```

**explanation**

- the distribution shows how class 0 has larger values, which aligns with the result of the previous box plot
- class 1 has a smaller variation
- some values overlap in both classes

```python
# split to X and Y
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# train model
from sklearn.svm import SVC

model = SVC(kernel='rbf')  
model.fit(X_train, y_train)
```

**explanation**

- this is the step where we train the data
- we use `rbf` since the data is not linear as shown in plot, and there are some overlaps. `rbf` creates non-linear boundaries

```python
# prediction
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

**explanation**

- `accuracy_score(y_test, y_pred)` measures the accuracy
- `classification_report(y_test, y_pred)` gives detail metrics
- A result of 85% accuracy is shown, indicating a good perfomance
- The precision, recall, f1-score, and support for both classes are high and balanced

```python
# viz
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

**explanation**

- The confusion matrix showed a total of 153 correct predictions, which is highly accurate
- Class 0 shows slightly more error but it's still well balanced
- 16 kecimen misclassified and 11 bensi misclassified

```python
# cross validation
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_scaled, y, cv=5)

print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

**explanation**

- `cv=5` splits data into 5 folds
- model is tested 5 times
- the cross-validation score is the outcome of each test
- average performance over 5 test is 87%, which shows high accuracy.

```python
for k in [3, 5, 10]:
    scores = cross_val_score(model, X_scaled, y, cv=k)
    print(f"k={k}, Mean Accuracy={scores.mean()}")
```

**explanation**

- increasing K to evaluate SVM with different number of folds
- As results showed, the best performance is at k=5, but results are all very closely aligned meaning the model is stable

```python
# try 50 50 split now
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X_scaled, y, test_size=0.5, random_state=42
)

model2 = SVC(kernel='rbf')
model2.fit(X_train2, y_train2)

y_pred2 = model2.predict(X_test2)

print("Accuracy (50-50):", accuracy_score(y_test2, y_pred2))
```

**explanation**

- the 50-50 split resulted in an accuracy score of 0.87, where as 80-20 resulted in a score of 0.85
- 50-50 split performed slightly better meaning more training data did not lead to a better result
- the larger test data may provide a more reliable model performance
- Both splits still has a very close result, indicating the model performs well overall.
