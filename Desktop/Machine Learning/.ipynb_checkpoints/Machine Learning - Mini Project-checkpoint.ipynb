**3. separate X and Y**

```python
X_train = train.drop(columns=["default"])
y_train = train["default"]

X_val = val.drop(columns=["default"])
y_val = val["default"] 
```

```python
print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)

print(X_train.head())
print(y_train.value_counts(normalize=True))
```

**4. clean**

**explanatin**

- The assignment says there are deleted values to simulate dirty data, but there are no missing values. I'm assuming the missing values were replaced as zeros.
- inspecting to see which rows has an unusual amount of zeros compared to existing data

```python
zero_pct = (X_train == 0).mean().sort_values(ascending=False)
print(zero_pct.head(10))
```

```python
high_zero_cols = zero_pct[zero_pct > 0.5]  # >50% zeros
print(high_zero_cols)
```

```python
print(X_train['A30'].describe())
print(X_train['A30'].value_counts().head(10))
```

**explanation**
- A30 has a high number of zeros. the zero in here looks like a placeholder for missing values since other existing values are larger

- A23 was also seen has a column with a lot of zeros, but when looking at the data, the other numbers in A23 are very close to zero, so it's not considered as an "unusual zeros column". 

```python
import numpy as np

# convert fake missing
X_train['A30'] = X_train['A30'].replace(0, np.nan)
X_val['A30'] = X_val['A30'].replace(0, np.nan)

# impute with median (from train only)
median_A30 = X_train['A30'].median()

X_train['A30'] = X_train['A30'].fillna(median_A30)
X_val['A30'] = X_val['A30'].fillna(median_A30)
```

**explanation**

- replaced the fake zeros with nan and fill it in with median value of the column
- feature engineering: Feature engineering involved identifying columns with suspicious zero values (e.g., A30) and treating them as missing. These values were imputed using the median of the training data to preserve distributional properties.

```python
# scale feature
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
# starting with logistic regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced', max_iter=1000)

model.fit(X_train_scaled, y_train)

y_prob = model.predict_proba(X_val_scaled)[:, 1]
```

**explanation**

- Starting with logistic regression as mentioned in the instructions. using standard scaler to scale features first
- from earlier inspection, it is shown that the data is very imbalanced
  - default:
  - 0    0.957447
  - 1    0.042553
  - almost 96% of the data is 0 (non-default), the model would lean towards predicting 0 all the time.
- using `class_weight = balanced` to balance the data, it gives class 1 a larger weight than class 0.

```python
# save first file

pd.DataFrame(y_prob).to_csv("results1.csv", index=False, header=False)
```

```python
# next model

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)
```

```python
rf.fit(X_train, y_train)
```

```python
y_prob_rf = rf.predict_proba(X_val)[:, 1]
```

**explanation**

- using random forest for the second model.
- `n_estimatos` uses 300 trees to vote for the final decision
- also using class_weight balanced to balance the data
- using ChatGPT, it suggested using n_jobs = -1 to speed up the training process. it still works without it, but makes it more efficient
- Using random forest since logistic regression finds the linear relationship, and random forest can capture more complex relations. Random forest is also great for classifications

```python
# save second file

pd.DataFrame(y_prob_rf).to_csv("results2.csv", index=False, header=False)
```
